```python
import pandas as pd

# Path to the betting_data.xlsx file in Colab
excel_path = '/content/betting_data.db.xlsx'  # Update with your actual file path

# Read the data into a Pandas DataFrame
df = pd.read_excel(excel_path)

# Display the first few rows of the dataset
print("First few rows of the dataset:")
print(df.head())

# Display basic statistics of the dataset
print("\nBasic statistics of the dataset:")
print(df.describe())

# Check for missing values
print("\nMissing values in the dataset:")
print(df.isnull().sum())

# Data Cleaning (if needed)
# For example, you might want to fill missing values or handle outliers here

# Summary Statistics on Profitability
total_sales = df['BetAmt'].sum() + df['BonusBetAmt'].sum()
total_revenue = df['WinAmt'].sum() + df['BonusWinAmt'].sum()
total_tax_paid = df['TaxAmt'].sum()
profit = total_revenue - total_tax_paid
profit_margin = (profit / total_sales) * 100

# Print Summary Statistics
print("\nSummary Statistics on Profitability:")
print(f"Total Sales: {total_sales}")
print(f"Total Revenue: {total_revenue}")
print(f"Total Tax Paid: {total_tax_paid}")
print(f"Profit: {profit}")
print(f"Profit Margin: {profit_margin:.2f}%")
```
